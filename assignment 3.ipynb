```python
# Import pandas for dataset handling
import pandas as pd
```

```python
# Load dataset
dataset = pd.read_csv("/Users/navya2005/Downloads/insurance_pre.csv")

# Display dataset
dataset
```

```python
# Convert categorical columns (sex, smoker) into numeric
dataset = pd.get_dummies(dataset, drop_first=True)

# Display processed dataset
dataset
```

```python
# Display column names
dataset.columns
```

```python
# Select independent variables
X = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]

# Display input data
X
```

```python
# Select dependent variable
Y = dataset[['charges']]

# Display output data
Y
```

```python
# Import train-test split
from sklearn.model_selection import train_test_split

# Split data (70% train, 30% test)
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.30, random_state=0
)

# Check sizes
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Scale features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Display scaled data (first few rows)
X_train[:5]
```

```python
# Import AdaBoost Regressor
from sklearn.ensemble import AdaBoostRegressor

# Create model with hyperparameters
model = AdaBoostRegressor(
    random_state=0,
    n_estimators=50,
    learning_rate=1.0,
    loss='linear'
)

# Train model (ravel fixes warning)
model.fit(X_train, Y_train.values.ravel())

# Display model
model
```

```python
# Show estimator weights
model.estimator_weights_
# Show estimator errors
model.estimator_errors_
```

```python
# Predict test data
Y_pred = model.predict(X_test)

# Display predictions
Y_pred
```

```python
# Import evaluation metric
from sklearn.metrics import r2_score

# Calculate R2 score
r2 = r2_score(Y_test, Y_pred)

print("R2 Score:", r2)
```

```python
# Save model using pickle
import pickle

filename = "finalized_model_AdaBoost.sav"
pickle.dump(model, open(filename, "wb"))

print("Model saved successfully")
```

```python
# Load saved model
loaded_model = pickle.load(open("finalized_model_AdaBoost.sav", "rb"))

print("Model loaded successfully")
```

```python
# Predict insurance charges
# Format: [age, bmi, children, sex_male, smoker_yes]

# IMPORTANT: Apply scaling before prediction
input_data = sc.transform([[18, 31.92, 0, 0, 0]])

result = loaded_model.predict(input_data)

print("Predicted Charges:", result)
```
